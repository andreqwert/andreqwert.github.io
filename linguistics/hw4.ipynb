## Домашнее задание 4

Две команды проводили матч по бегу. Итоги соревнований определялись следующим образом: если $k$-ый по силе участник одной из команд обогнал такого же по силе участника другой команды, то этой команде присваивалось очко; если же они показали одинаковые результаты, то по 0,5 очка давалось каждой из команд. Гарантируется, что в обеих командах одинаковое количество человек. Напишите программу, получающую на вход две строчки результатов в формате $mm:ss$, разделённых пробелом, и возвращающую счёт матча.

```python
k = int(input('Число участников: '))

c1, c2 = 0, 0
for i in range(k):
    s1 = input('Время бегуна 1-ой команды: ')
    s2 = input('Время бегуна 2-ой команды: ')
    print('--------------------------------')
    s1_time = int(s1.split(':')[0]) * 60 + int(s1.split(':')[1])
    s2_time = int(s2.split(':')[0]) * 60 + int(s2.split(':')[1])
    if s1_time > s2_time:
        c2 += 1
    elif s1_time == s2_time:
        c1 += 0.5
        c2 += 0.5
    else:
        c1 += 1
print('{}:{}'.format(c1, c2))
```
