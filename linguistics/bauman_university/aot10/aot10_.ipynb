## 1. Предобработка входных данных

```python
from pymystem3 import Mystem
m = Mystem()

WORD = "билет"
window_size = 2 # размер окна для анализа


def get_tags(context):
    """Каждому слову присваиваем свою часть речи"""
    
    ws = m.analyze(context)
    tags = []
    for w in ws:
        try:
            word = w['text']
            tag = w['analysis'][0]['gr'].split(',')[0].split('=')[0]
            tags.append([word, tag])
        except:
            pass
    return tags


def build_tagged_sentences(context):
    """Строим пары лемма:часть_речи для каждого предложения контекста"""
    
    lemmas, tags = [], []
    tagged_sentences = []
    sentences = context.split('.') # Делим все предложения контекста по точке
    for sentence in sentences:
        tagged_sentence = get_tags(sentence) # Получаем частиречные теги на каждое слово предложения
        lemmas = [m.lemmatize(word)[0] for word, tag in tagged_sentence] # Каждое слово предложения лемматизируем
        tags = [tag for word, tag in tagged_sentence]
        tagged_sentences.append(list(zip(lemmas, tags))) # Составляем пару лемма:часть_речи
    return tagged_sentences[0:len(sentences)-1] # так как split по точке, то получилось, что последнее предложение - пустое: удаляем его


def get_window_words(sentence):
    """
    Вычисляем, какие слова будут играть роль признаков. 
    Ими будут ±2 слова (ширина окна window_size=2) от слова WORD, 
    семантическое значение которого мы ищем.  
    """
    
    window_words = []
    for word_location in range(len(sentence)):
        if (sentence[word_location])[0] == WORD:
            
            """Это условие позволит обрабатывать предложения, в которых слово WORD стоит в самом начале"""
            if word_location < window_size:
                window_words.append(sentence[0:word_location+window_size+1])
            else:
                window_words.append(sentence[(word_location-window_size):(word_location+window_size+1)])
    
    return window_words
```

По сути получившееся окна - это и есть вектора признаков.    
Теперь надо вытащить:
- Сами слова;
- Части речи для каждого из слов окна.

```python
import numpy as np
import pandas as pd

NUM_OF_SENTENCES = 9 # количество предложений для каждого контекста
NUM_OF_WORDS_PER_SENTENCE = window_size * 2 # число слов слева и справа от искомого
NUM_OF_TAGS_LEFT = NUM_OF_TAGS_RIGHT = window_size # число признаков-частиречных_тегов слева и справа от центрального элемента


def make_dataframe(context_num):
    """Создаём (пока что) пустой датафрейм заданных размеров"""
    
    features = []
    for i in range(1, (NUM_OF_WORDS_PER_SENTENCE+2)):
        features.append(f'f{i}_word')
    for j in range(1, (NUM_OF_TAGS_LEFT+1)):
        features.append(f'f{j}_left_tag')
    for k in range(1, (NUM_OF_TAGS_RIGHT+1)):
        features.append(f'f{k}_right_tag')
    
    data = {}
    for feature in features: # создаём столбцы с признаками
        data[feature] = np.zeros(NUM_OF_SENTENCES)
    data['class_TARGET'] = [context_num]*NUM_OF_SENTENCES # создаём столбец с целевой меткой (классом документа)
    df = pd.DataFrame(data, index=None)
    return df
```

```python
def fill_dataframe(sentences, df):
    """Заполнение датафрейма значениями"""
    
    indexes = []
    for sentence in sentences:
        for word in sentence:
            if word[0] == WORD: # если слово - это WORD, то мы записываем его как серединное, а потом просто от него
                                # ставим слова слева и справа (при наличии таковых), и так пока все ±2 не закончатся
                df['f3_word'] = word[0]
                indexes.append(sentence.index(word)) # индекс слова WORD


    """            
    Мы дополняем None'ами те ячейки предложения, которые входят в окно, но в которых нет слов.
    Это делается для того чтобы привести все предложения к одной размерности.

    Например, в этом предложении для окна 2 относительно слова "билет" нет первого слова предл-я -> делаем первое слово пустым:
    " _ Проездные билеты, посвященные стодесятилетию " <- заместо пропуска "_" должен стоять None
    """
    df_sentences = []
    for index, sentence in zip(indexes, sentences):
        if index < window_size:
            # Если в окне из 5 слов предложение начинается сразу со слова WORD, то заполняем первую и вторую ячейку датафрейма None'ами
            if index == 0:
                sentence.insert(0, ('None', 'None'))
                sentence.insert(1, ('None', 'None'))
                index -= 1
            # Если в окне из 5 слов предложение содержит WORD вторым словом, то первую ячейку заполняем None'ом.
            elif index == 1:
                sentence.insert(0, ('None', 'None'))
                index -= 1
            else:
                index -= 1
        while len(sentence) < NUM_OF_WORDS_PER_SENTENCE+1:
            sentence.append(('None', 'None'))
        df_sentences.append(sentence)
    
    
    """Заполним датафрейм получившемися значениями"""
    for i in range(1, window_size*2+1+1): # потому что всего 5 колонок со словами
        # первый индекс вытаскивает только первые слова из предложений,
        # второй индекс вытаскивает только первое слово из пары "слово, часть речи"
        df[f'f{i}_word'] = pd.Series([word[i-1][0] for word in df_sentences])

    for j in range(1, NUM_OF_TAGS_LEFT+1):
        df[f'f{j}_left_tag'] = pd.Series([word[j-1][1] for word in df_sentences])

    for k in range(window_size-1, NUM_OF_TAGS_RIGHT+1):
        df[f'f{k}_right_tag'] = pd.Series([word[window_size+k][1] for word in df_sentences])
    
    return df   
```

```python
def build_dataframe_from_context(context, context_num):
    """ГЛАВНАЯ ФУНКЦИЯ ПРЕДОБРАБОТКИ: СОЗДАНИЕ ДАТАФРЕЙМА ИЗ КОНТЕКСТА"""
    
    sentences = []
    tagged_sentences = build_tagged_sentences(context)
    for sentence in tagged_sentences:
        sentences.append(get_window_words(sentence)) 
    empty_resized_df = make_dataframe(context_num)
    sentences = [sentence[0] for sentence in sentences]
    
    df = fill_dataframe(sentences, empty_resized_df)
    return df
    

df_context1 = build_dataframe_from_context(context1, 1)
df_context2 = build_dataframe_from_context(context2, 2)
```

Для кодирования признаков воспользуемся [One-hot encoder'ом](https://dyakonov.org/2016/08/03/python-%D0%BA%D0%B0%D1%82%D0%B5%D0%B3%D0%BE%D1%80%D0%B8%D0%B0%D0%BB%D1%8C%D0%BD%D1%8B%D0%B5-%D0%BF%D1%80%D0%B8%D0%B7%D0%BD%D0%B0%D0%BA%D0%B8/). 
Для построения модели Байесовского классификатора все признаки должны иметь числовой тип, поэтому мы будем подавать на вход OHE всё, кроме целевой метки.

```python
"""Применяем непосредственно OHE"""
from sklearn.preprocessing import OneHotEncoder

columns_to_transform = ['f1_word', 'f2_word', 'f4_word', 'f5_word', 
                        'f1_left_tag', 'f2_left_tag', 'f1_right_tag', 'f2_right_tag']


def code_myohe(df, feature):
    for i in df[feature].unique():
        df[feature + '=' + i] = (df[feature] == i).astype(float)
    return df
    

for column_feature in columns_to_transform:
    code_myohe(df, column_feature)
df = df.iloc[:, 8:]  # !!! Тут надо поменять номера колонок, чтоб добавить или убрать столбцы с частиречными тегами
                     # Чтобы не брать частиречные теги, надо ограничить iloc до [8:-7]
# В получившейся таблице первый столбик - целевая метка, остальное - признаки.
print("ИСПОЛЬЗУЕМЫЕ ПРИЗНАКИ: ")
print(df.columns.values[1:]) # [1:] - потому что в значащие признаки целевая метка не включается
print('Количество признаков: ', len(df.columns.values[1:]))
```

Итоговый датафрейм сверху содержит целевую метку (в первом столбце, по аналогии - это $D_1$) и признаки документа (все остальные столбцы, по аналогии - $X_1, X_2, .., X_n$).

Переименуем для удобства:

```python
def rename_columns(df):
    df = df.rename(index=str, columns={"class_TARGET": "context_num"})
    for feature_num, column_name in zip(range(1, len(df.columns.values)), list(df.columns.values[1:])):
        df.rename(index=str, columns={column_name:f'X{feature_num}'}, inplace=True)
    return df

df = rename_columns(df)
```

```python
def get_headers(df):
    # Получаем header'ы (=названия_столбцов) датафрейма
    return np.array(df.columns.values[1:])

headers = get_headers(df)
```

```python
def get_classes_array(df):
    # Получаем вектор с целевыми метками
    classes = ['C'+str(i) for i in np.array(df['context_num'])]
    return classes

classes = get_classes_array(df)
```

```python
def get_features(df, headers):
    # Вытаскиваем из датафрейма по значению 1.0 необходимые header'ы из множества header'ов.
    # То есть если слову соответствует единичка, то мы смотрим на название его столбца и записываем это название в качестве признака
    features_all = []
    for row in df.itertuples():
        features_for_one_sentence = []
        for column_name, value in zip(headers[0:], row[2:]):
            if value == 1.0:
                features_for_one_sentence.append(column_name)
        features_all.append(features_for_one_sentence)
    return features_all

features_all = get_features(df, headers)
```

```python
from collections import Counter


def define_Nxc(train_data):
    """Найдем, сколько раз встречается признак X_i при условии класса C_j"""
    
    Nxc = {}
    c1_fs, c2_fs = [], []
    for feature, class_ in zip(np.array(train_data['Features']),
                               np.array(train_data['Classes'])):
        if class_ == 'C1':
            c1_fs += feature
        elif class_ == 'C2':
            c2_fs += feature

    c1_fs, c2_fs = np.array(c1_fs), np.array(c2_fs)
    Nxc['C1'] = Counter(c1_fs)
    Nxc['C2'] = Counter(c2_fs)
    #print('Nxc: ', Nxc)
    return Nxc



def define_Nc(train_data):
    """Найдём общее число неуникальных признаков, принадлежащих каждому из классов.
    Для этого здесь создаётся словарь, ключами которого являются классы, а значениями - количество неуникальных признаков"""
    
    Nc = {}
    c1_fs, c2_fs = [], []
    for feature, class_ in zip(np.array(train_data['Features']),
                               np.array(train_data['Classes'])):
        if class_ == 'C1':
            c1_fs += feature
        elif class_ == 'C2':
            c2_fs += feature
            
    Nc['C1'] = len(c1_fs)
    Nc['C2'] = len(c2_fs)
    #print('Nc: ', Nc)
    return Nc
    
    
    
def define_k(train_data):
    """Найдём суммарное число признаков обучающей выборки"""
    sum_feature_vec = []
    for feature_vec in train_data['Features']:
        for feature in feature_vec:
            sum_feature_vec.append(feature)
    #print('k =', len(set(sum_feature_vec)))
    return len(set(sum_feature_vec))
        
           
        
def apply_multinomial_bayes_on_train_data(train_data):
    Nxc = define_Nxc(train_data)
    Nc = define_Nc(train_data)
    k = define_k(train_data)
    
    
    print('ВЕРОЯТНОСТИ ПО БАЙЕСОВСКОМУ КЛАССИФИКАТОРУ: ')
    Xs = []
    Ps_C1, Ps_C2 = [], []
    data_probs = {}
    for ci, xi in zip(Nxc.keys(), Nxc.values()):
        if ci == 'C1':
            for i in range(1, k+1):
                P = (xi[f'X{i}'] + 1) / (Nc[ci] + k)
                Ps_C1.append(P)
            data_probs[ci] = Ps_C1

        elif ci == 'C2':
            for i in range(1, k+1):
                P = (xi[f'X{i}'] + 1) / (Nc[ci] + k)
                Ps_C2.append(P)
            data_probs[ci] = Ps_C2
    
    # Добавляем для поиска столбец с пронумерованными названиями признаков
    for i in range(1, k+1):
        Xs.append('X' + str(i))
    data_probs['Xi'] = Xs
    
    df_probs = pd.DataFrame(data_probs, index=None)
    print(df_probs, '\n\n')
    return df_probs


def apply_multinomial_bayes_on_test_data(df_train, test_data):
    
    """
    Достаём тестовую выборку и для каждого предложения смотрим на вероятность принадлежность признака к классу. 
    Оставляем в исходном датафрейме, определённом функцией выше, только те признаки, на которые указывает вектор
    признаков тестовых предложений.
    
    
    ЧТОБЫ СЧИТАЛОСЬ БЕЗ ЧАСТИРЕЧНЫХ ТЕГОВ: надо просто убрать колонки с частиречными бинарными признаками в датафреймах на обучение и тест.
    Для этого нужно поменять диапазоны обучающей выборки на df.iloc[:, 8:-7] (потому что последние 7 колонок относятся к признакам-тегам, а первые 8 
    признаков вообще небинарные и они не нужны) и тестовой выборки на df_test.iloc[:, 8:-7]
    """
    for sentence, test_sentence_itself in zip(range(1, len(test_data)+1), test_sentences):
        test_features_vec = test_data.iloc[(sentence-1):sentence, :]['Features']
        print(f'\n\n-------{test_sentence_itself.strip()}-------\n') # strip() - чтоб удалялись пробелы слева и справа от предложения
        res = calculate_more_prior_class(df_probs, test_features_vec)
    return None


def calculate_more_prior_class(df_probs, test_feature_vec):
    
    print('С ЧАСТИРЕЧНЫМИ ТЕГАМИ: ')
    features_probs_C1, features_probs_C2 = [], [] # Списки для вероятностей принадлежности к классам С1 и С2.
    test_feature_vec = np.array(test_feature_vec)
    print('Признаки тестовой выборки: ', test_feature_vec)
    for features in test_feature_vec:
        for feature in features:
            # тут складываются единичные вектора вероятностей, в каждом векторе - по одной вероятности. Изъяны pandas.
            features_probs_C1.append(sum(df_probs.loc[df_probs['Xi'] == feature, :]['C1']))
            features_probs_C2.append(sum(df_probs.loc[df_probs['Xi'] == feature, :]['C2']))
    
    # 0.5 - потому что вероятность самогО класса С1 такая же, как вероятность класса С2
    # Это потому что у нас сбалансированная обучающая выборка: в каждой по NUM_OF_SENTENCES предложений
    P_C1 = 0.5 * np.prod(features_probs_C1)   
    P_C2 = 0.5 * np.prod(features_probs_C2)
    
    if P_C1 > P_C2:
        relative_prob = 100 * (P_C1 / (P_C1 + P_C2))
        print('Билет в значении ЭКЗАМЕНАЦИОННЫЙ. P = {0:.5f}%'.format(relative_prob))
        print('Билет в значении ПРОЕЗДНОЙ. P = {0:.5f}%'.format(100 - relative_prob))
    else:
        relative_prob = 100 * (P_C2 / (P_C1 + P_C2))
        print('Билет в значении ЭКЗАМЕНАЦИОННЫЙ. P = {0:.5f}%'.format(100 - relative_prob))
        print('Билет в значении ПРОЕЗДНОЙ. P = {0:.5f}%'.format(relative_prob))

    
            
df_probs = apply_multinomial_bayes_on_train_data(train_data)
df_test = apply_multinomial_bayes_on_test_data(df_probs, test_data)
```
