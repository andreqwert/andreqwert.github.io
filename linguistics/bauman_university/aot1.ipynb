## Задание №1: 
- Взять художественное произведение;
- Извлечь слова;
- Посчитать частоты слов;
- Проверить закон Ципфа.

```python
# В качестве художественного произведения был выбран Шекспировский корпус

corpus_name = 'Three_Men_in_a_Boat.txt'
with open(corpus_name) as fin:
    corpus = fin.read()
```

**Применим библиотеку *re* для работы с регулярными выражениями. Она необходима для того, чтобы предложения можно было разделять методом *split* не только по пробелам, но и по другим символам. Иначе получатся отдельные "слова", состоящие из знаков препинания, что неприемлимо.**

```python
import re
words = re.split("[ \n;,'`.?!:']", corpus)
print(words[:100])
```

**Избавимся от пустых строк при помощи функции *filter*.**

```python
words = list(filter(None, words)) #None - потому что фильтрует пустые строки
print(words[:100])
```

**Отсортируем все слова в лексигографическом порядке**:

```python
words.sort()
print(words[:100])
```

**Добавляем слова в словарь и смотрим на то, сколько раз встретил то или иное слово в корпусе**

```python
from collections import defaultdict, Counter

word_counts = defaultdict(int) #Определяем, что в key:value словаря значение value - целое число

#Пробегаем по всем словам в списке words и с помощью словаря сопоставляем каждому из слов - количество раз, которое оно было написано
for each_word in words:
    word_counts[each_word] += 1
```

**Напечатаем 20 самых частых слов при помощи функции *Counter***:

```python
import pprint as pp

most_frequent_words = Counter(word_counts).most_common(500)
print('\nThe most frequent words of .txt with the numbers of entrances:\n')

pp.pprint(most_frequent_words)
```

**Проверим закон Ципфа. По нему частота $n$-ого слова в списке самых частых слов должна быть обратно пропорциональна его порядковому номеру $n$**:

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.figure(figsize=(9, 6))
#Получим порядковые номера слов
nums = np.arange(1, 501, 1) #от 1 до 500 включительно

#Извлечем частоты слов
words_freqs = []
for key, value in most_frequent_words:
    words_freqs.append(value)
#Построим график
plt.title('Закон  Ципфа')
plt.grid() #чтоб была линия сетки
plt.plot(nums, words_freqs)
plt.xlabel('Ранг слов')
plt.ylabel('Частота слов')
```
