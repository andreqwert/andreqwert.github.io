## Задание 4

- Выбрать и обработать произвольный текст морфологическими анализаторами;
- Сделать частотные списки лемм;
- Выдать леммы по мере снижения их частот.

```python
# В качестве текста был выбран кусок произведения "Понедельник начинается в субботу" Стругацких
corpus_name = 'strugatskie.txt'
with open(corpus_name) as fin:
    corpus = fin.read()

# Убираем знаки препинания
import re
corpus = re.split("[ \n«»;,'`—.?!:']", corpus)

# Убираем пустые строки
corpus = list(filter(None, corpus)) #None - потому что фильтрует пустые строк
corpus = ' '.join(corpus)
```

```python
from pymystem3 import Mystem
```

```python
mystem = Mystem()
```

```python
# Обработка для получения лемм
lemmas = ''.join(mystem.lemmatize(corpus))
```

```python
# Частотные списки лемм
lemms = {}
for word in lemmas.strip().lower().split():
    lemms[word] = lemms.get(word, 0) + 1 
print(lemms)
```

```python
# Выдаем леммы по мере снижения их частот
from operator import itemgetter
sorted(lemms.items(), key=itemgetter(1), reverse=True)
```
